# Scalecast ðŸŒ„ with eCommerce Hourly

This notebook details a timeseries analysis and forecasting application performed with scalecast. The notebook starts with an exploratory data analysis; moves to timeseries decomposition; forecasats manually with an ARIMA model; forecasts manually with a multiple linear regression model; then moves to automated forecasting with scikit-learn models, Facebook Prophet, and LinkedIn Greykite/Silverkite; finally finishing with combination modeling.

The utilized dataset is available on kaggle: https://www.kaggle.com/carrie1/ecommerce-data/

## Library Imports
First, let's import the libraries and read the data. Some data preprocessing in pandas will be necessary before getting scalecast involved.

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from scalecast.Forecaster import Forecaster
from scalecast import GridGenerator

plot_dim = (15,7)
sns.set(rc={'figure.figsize':plot_dim})
```

It is important to let pandas know that the date column should be datetime type! That's why we pass the argument `parse_dates=['InvoiceDate']`

```python
data = pd.read_csv('eCommerce.csv',parse_dates=['InvoiceDate'])
```

## Exploratory Data Analysis

Let's view the data's first five rows.

```python
data.head()
```

Now, let's view the data's dimensions.

```python
data.shape
```

Let's see how much time these 550,000 spans.

```python
print('first date in data:',data['InvoiceDate'].min())
print('last date in data:',data['InvoiceDate'].max())
```

In spite of there being over half a million rows, there is only about a year's worth of data to analyze, which means we could be dealing with a dataset that recorded every sale in that timeframe. Before proceding, we should decide a datetime frequency to aggregate the data to, otherwise we will be trying to forecast with an incosistent interval of time between observations. This decision depends on what question you are trying to answer. For this example, let's try answering the question of if we can predict hourly gross sales quantity over a week's worth of time. This means we will be removing sales that were negative in value (probably representing returns) and aggregating the entire dataframe to the hourly level. Although, as we will see, the scalecast package is probably better-suited to forecast daily sales with this dataset.

```python
dt_aggr = 'D'
data = data.loc[data['Quantity'] > 0]
data['DateTime'] = data['InvoiceDate'].dt.floor(dt_aggr)
tbl = data.groupby('DateTime')['Quantity'].sum().reset_index()
tbl.head()
```

```python
tbl.shape
```

```python
print('first date in data:',tbl['DateTime'].min())
print('last date in data:',tbl['DateTime'].max())
```

It is possible that after making this aggregation to the hourly level, some date observations in the hourly range are missing. Some forecasting libraries will process missing data for you automatically, but because scalecast mixes so many model concepts, it is necessary to have every possible date in a given range represented. If we run the code below, we can check to see if the shape of a dataframe with every possible date between our data's start and end is the same size (row-wise) as our data. If it is, we don't have to do anything else.

```python
all_dates = pd.DataFrame({'DateTime':pd.date_range(start=tbl['DateTime'].min(),end=tbl['DateTime'].max(),freq=dt_aggr)})
all_dates.shape
```

```python
full_data = all_dates.merge(tbl,on='DateTime',how='left').fillna(0).sort_values(['DateTime'])
full_data.plot(x='DateTime',y='Quantity',title='eCommerce Sales Quantity Original')
plt.show()
```

Because there are more observations in the `all_dates` dataframe, we have to merge the two dataframes together using a left join. We fill in missing values in the `tbl` dataframe with 0 to represent no sales in the timeframes that the dates are missing. We can then plot the entire series.

Because we notice two ourlier when plotting we can fill in those outlier with the original data's median value.

Note, there are more than 3x as many dates with 0 sales as dates with non-0 sales. Because of this, predicting this variable at the hourly level may not be most effective. There is another notebook in this repository that performs this anlysis at the daily level.

```python
full_data = all_dates.merge(tbl,on='DateTime',how='left').fillna(0).sort_values(['DateTime'])
full_data.plot(x='DateTime',y='Quantity',title='eCommerce Sales Quantity Original')
plt.show()

# fill in outliers
full_data.loc[full_data['Quantity'] == full_data['Quantity'].max(),'Quantity'] = tbl['Quantity'].median()
full_data.loc[full_data['Quantity'] == full_data['Quantity'].max(),'Quantity'] = tbl['Quantity'].median()
full_data.plot(x='DateTime',y='Quantity',title='eCommerce Sales Quantity Outliers Removed')
plt.show()
```

Much better. Let's now try to get a better idea of how the data is distributed.

```python
full_data.describe()
```

```python
full_data['Quantity'].hist(bins=25)
plt.show()
```

It looks like we have a strong right skew in the data.

## Forecast with Scalecast
To load the object, we call the Forecaster() function with the `y` and `current_dates` parameters specified. If we return the object at that time, it will give us general information about object's current structure. Contributions needed to improve the information returned when the object is called.

```python
f = Forecaster(y=full_data['Quantity'],current_dates=full_data['DateTime'])
f
```

The first thing you should do after initializing the object is set its test length. What that length is is up to you. The longer the length, the more confident you can be about your reported error/accuracy metrics. The library requires a test length of at least 1. Let's use 4 weeks of data, which is 4x greater than our selected forecast period.

```python
f.set_test_length(31)
```

Before beginning the forecasting process, we should get a better idea of the signals within the time series itself. Using ACF, PACF, and Periodogram plots, we can observe how the series is auto-correlated. We leave the test set out of all visualizations (`train_only=True`) to not leak data when making decisions about which signals exist in the data.

```python
f.plot_acf(train_only=True,lags=60)
plt.show()
```

```python
f.plot_pacf(train_only=True,lags=60)
plt.show()
```

```python
a, b = f.plot_periodogram(diffy=True,train_only=True)
plt.semilogy(a, b)
plt.show()
```

We clearly see that there is strong autocorrelation back at least 24 periods (one full day). There is probably some daily seasonilty as well. We can further visually decompose the series by using the function below:

```python
f.seasonal_decompose(train_only=True).plot()
plt.show()
```

Now, let's test the data's stationarity using the Augmented Dickey-Fuller test. The null hypothesis of this test is that the data is not stationary. To return the p-val and critical value from this test, pass `full_res=True` as an argument. The way we have the test specified, it will simply print out the implications from the test.

```python
isstationary = f.adf_test(quiet=False)
```

```python
f.plot_acf(diffy=1,train_only=True,lags=60)
plt.show()
```

```python
f.plot_pacf(diffy=1,train_only=True,lags=60)
plt.show()
```

```python
f.seasonal_decompose(diffy=1,train_only=True).plot()
plt.show()
```

```python
f.plot_pacf(train_only=True,lags=60)
plt.show()
```

The results of this test suggest the series is stationary. If it wasn't, we could call the `diff()` function to attempt to make it stationary. Many analysts avoid differencing time series when forecasting because of the difficulty of producing usable results. With the scalecast package, it is extremely easy. Some of the other examples in the repository go over this.

### ARIMA
Now, let's run a common time-series model: ARIMA. Using the output from the plots above, as well as the results from the ADF test, we can specify a 3,0,0 x 2,0,0 ordered model. The seasonal period will be one day--24 periods.

Before running any forecast, we need to generate a forecast period. We are attempting to predict a week of hourly sales, so we set a forecast period of 1 week.

```python
f.generate_future_dates(31)
```

```python
f.set_estimator('arima')
f.manual_forecast(order=(6,1,0),seasonal_order=(2,0,0,7),trend='t')
```

We can view the results of the model by plotting the test-set predictions with the actual test-set observations.

```python
f.plot_test_set()
```

Let's now view the forecast into unknown periods by calling the `plot()` function. We can also print out interesting model information using the `print_attr` argument. Let's print out the Test-Set RMSE. We can call this with either 'LevelTestSetRMSE' or 'TestSetRMSE'. These will only be different when the series was differenced before modeling. For our example, these will return identical values.

```python
f.plot(print_attr=['LevelTestSetRMSE'])
```

The RMSE from this model can be considered a baseline to beat with more advanced modeling.

Most models allow feature information to be saved from them using `save_feature_importance()` or `save_summary_stats()`. Some models don't allow either. We can see the summary statistics for an ARIMA model by saving and then exporting them to a dataframe. Note that the ARIMA model is saved to the object as 'arima' by default, but if you want to use several arima models, you can differentiate them using the `call_me` argument when calling `manual_forecast()`.

```python
f.save_summary_stats()
```

```python
f.export_summary_stats(model='arima')
```

### Linear Regression
We can use many other models through scalecast. One of the most basic of these is Multiple Linear Regression. Unlike ARIMA, we don't specify orders on this model, but we can add similar regressors, including autoregressive and seasonal terms, as well as time trends.

First, let's begin with autoregressive terms, which are lags of the dependent variable values. We can add 1-24 lags, as well as the 48th lag to account for some seasonality.

```python
f.add_ar_terms(6) # 1-6 lags
```

```python
f.add_AR_terms((2,7)) # 24th and 48th lags
```

```python
f.diff()
```

To outclass the ARIMA model, we can also add more seasonlity than just hourly. We can add the hour, day, week, dayofweek, and month regressors. We should specify them with sin/cos transformations to increase the cyclical nature of the signals we are trying to model, which will increase accuracy. `dummy=True` is also available as an argument to use dummy variables. If `raw=True` is used, the raw integer data will be used, but this is discouraged since many models will try to fit a linear relationship when such a relationship isn't ideal.

```python
f.add_seasonal_regressors('day','week','dayofweek',raw=False,sincos=True)
```

We can also add a time trend to our modeling process.

```python
f.add_time_trend()
```

Other options for adding regressors include:
- `add_combo_regressors()`
- `add_poly_terms()`
- `add_covid19_regressor()`
- `add_other_regressor()`
- `ingest_Xvars_df()`

Let's see the names of the regressors we have added.

```python
f.get_regressor_names()
```

When adding these and other kinds of regressors, it is possible to change the names of some of them. This can come in handy to reference later. If any regressors begin with uppercase "AR", the forecasting mechanisms in the models that will be explored later will assume such terms are autoregressive in nature and those terms are handled differently. So, be careful when naming variables.

Let's move to modeling with a linear model. We can manually forecast such a model, since it doesn't have hyperparemeters, although it is possible to optimize the following arguments in any model sourced from scikit-learn:
- `Xvars` (arguments include "all", None, and list-like objects) -- default is always None, but for models that require regressors, this is treated the same as "all"
- `normalizer` (arguments are None, "minmax", "normalize", and "scale" -- default is always "minmax"

```python
f.set_estimator('mlr')
f.manual_forecast(normalizer=None,Xvars=None)
```

Like the ARIMA model, we can see its performance on the test set. Note that all forecasting in this module, even on the test set, is dynamic in nature so that no training-set information is leaked. We can trust that this is a true performance on three months of out-of-sample data.

```python
f.plot_test_set(level=True)
```

We see that it doesn't have the ability to pick up seasonal patterns in the data, unlike the ARIMA model that struggled to do so. However, it doesn't capture some of the more extreme values well at all.

Notice that the evaluation of this model takes a long time. That is because to properly process autoregressive terms, loops are used in the source code, which slows down model evaluation considerably, especially for large test/validation sets and large forecast intervals. Scalecast performs best on midsize datasets of about 100-2000 observations for that reason, so we are pushing its application to its feasable limit.

Let's now see the MLR forecast and test-set RMSE.

```python
f.plot(print_attr=['LevelTestSetRMSE'])
```

With scikit-learn models, such as MLR, permutation feature importance information from the [eli5](https://eli5.readthedocs.io/en/latest/blackbox/permutation_importance.html) package can be saved and exported to a dataframe, similar to saving summary stats from an ARIMA model. Each weight is the average decrease in accuracy when that particular variable is substituted for random values over 10 iterations, so the variables that cause the largest decreases are considered the most important. The output of the exported data is therefore ranked in terms of which variables are most-to-least important, according to this methodology.

```python
f.save_feature_importance()
```

```python
f.export_feature_importance(model='mlr')
```

### Elasticnet and Auto-Forecasting

To optimize the hyperparameters of and auto-forecast with models in scalecast, we use a grid-search approach on a validation set of data--a period of time before the test set. The grids are completely customizable, but standard template grids are available by calling the function below:

```python
GridGenerator.get_example_grids()
```

`GridGenerator.get_empty_grids()` is also available. The user would then open the created file (Grids.py) and fill in the empty dictionary with hyperparameter values that she wants to use. The example grids can usually be used for adequated performance, but contributions to improving the default values in these grids are welcome.

These grids are saved to your working directory as Grids.py. Their structure is that of `Dict[str:list-like]` and scalecast automatically knows how to look for them. You can also pass your own grid manually by using `ingest_grid()` and passing a `str` value, which corresponds to a namespace of a grid in the Grids.py file, or a grid `dict` type. By default, the code below will ingest the grid named elasticnet from the Grids.py file in this same directory.

You can also create really big grids and limit them randomly by calling `limit_grid_size(n:int|float)`. If `int`, must be >0 to denote the number of combinations in the grid to keep. If `float`, must be between 0 and 1 to represent a portion of the grid's original size to randomly keep. `random_seed` is available as an argument in this method and is None by default.

```python
f.set_validation_length(15)
f.set_estimator('elasticnet')
f.tune() # automatically imports the elasticnet grid from Grids.py
f.auto_forecast()
```

```python
f.plot_test_set()
```

```python
f.plot(print_attr=['LevelTestSetRMSE'])
```

```python
f.save_feature_importance()
```

```python
f.export_validation_grid(model='elasticnet').sort_values(['metric_value']).head(15)
```

```python
f.export_feature_importance(model='elasticnet')
```
